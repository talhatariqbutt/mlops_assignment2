```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extract(sources, max_articles_per_site=100):
    data = []
    for source_url in sources:
        print("Extracting data from:", source_url)
        reqs = requests.get(source_url)
        soup = BeautifulSoup(reqs.text, 'html.parser')
        articles = soup.find_all('article', class_='story')
        article_count = 0
        for article in articles:
            if article_count >= max_articles_per_site:
                break
            title_elem = article.find('h2', class_='story__title')
            if title_elem:
                title = title_elem.text.strip()
                link = title_elem.find('a')['href']
            else:
                title = None
                link = None
            
            description_elem = article.find('div', class_='story__excerpt')
            description = description_elem.text.strip() if description_elem else None
            
            data.append({'source': source_url, 'title': title, 'description': description, 'link': link})
            article_count += 1
            print("Articles extracted:", article_count)
    df = pd.DataFrame(data)
    return df


sources = ['https://www.dawn.com/latest-news']
df = extract(sources, max_articles_per_site=100)
df
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# URL of the page to scrape
url = "https://www.bbc.com/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find all elements with the specified class that contain articles
articles = soup.find_all(class_="sc-e5949eb5-1")

# Initialize lists to store the data
titles = []
links = []
descriptions = []

# Loop through each article and extract title, link, and description
for article in articles:
    # Extract title
    title = article.find(class_="sc-4fedabc7-3").get_text(strip=True)
    titles.append(title)
    
    # Extract link if it exists
    link_tag = article.find("a")
    if link_tag:
        link = "https://www.bbc.com" + link_tag.get("href")
    else:
        link = None
    links.append(link)
    
    # Extract description
    description = article.find(class_="sc-e5949eb5-0").get_text(strip=True)
    descriptions.append(description)

# Create a DataFrame to store the data
bbc_df = pd.DataFrame({
    "Title": titles,
    "Link": links,
    "Description": descriptions
})

# Display the DataFrame
print(bbc_df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin

def extract(sources, max_articles_per_site=300):
    data = []
    for source_url in sources:
        print("Extracting data from:", source_url)
        reqs = requests.get(source_url)
        soup = BeautifulSoup(reqs.text, 'html.parser')
        links = [link.get('href') for link in soup.find_all('a')]
        article_count = 0
        for link in links:
            abs_link = urljoin(source_url, link)  # Convert to absolute URL
            reqs = requests.get(abs_link)
            soup = BeautifulSoup(reqs.text, 'html.parser')
            articles = soup.find_all('article')
            for article in articles:
                if article_count >= max_articles_per_site:
                    break
                title = article.find('h3').text.strip() if article.find('h3') else None
                description = article.find('p').text.strip() if article.find('p') else None
                data.append({'source': source_url, 'title': title, 'description': description})
                article_count += 1
                print("Articles extracted:", article_count)
            if article_count >= max_articles_per_site:
                break
    bbc_df = pd.DataFrame(data)
    return bbc_df

sources = ['https://www.bbc.com/']
bbc_df = extract(sources, max_articles_per_site=300)

bbc_df
```

```python

def clean_text(text):
    if isinstance(text, str):
        # Remove special characters and digits
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        # Convert text to lowercase
        text = text.lower()
        return text
    else:
        return ''


def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word not in stop_words]
    return ' '.join(filtered_text)

def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    word_tokens = word_tokenize(text)
    lemmatized_text = [lemmatizer.lemmatize(word) for word in word_tokens]
    return ' '.join(lemmatized_text)

combined_df['description'] = combined_df['description'].apply(clean_text)
combined_df['description'] = combined_df['description'].apply(remove_stopwords)
combined_df['description'] = combined_df['description'].apply(lemmatize_text)

combined_df.to_csv('preprocessed_data.csv', index=False)

print("Preprocessing completed and saved to preprocessed_data.csv")
```
